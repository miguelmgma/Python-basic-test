### 1.1: crea una funcion a la cual le pases por parametros el numero de filas y el numero de columnas y te la rellene con numeros aleatorios del 1 al 100 con 2 decimales. 
Crea una matriz de 10 filas y 7 columnas. <font color="#9E1B32">**(No utilices NUMPY ni PANDAS en este apartado, en los demás puedes usarlo)**</font>

```python
import random

def mat_aleat(n_rows,n_cols):
    rows = []
    for r in range(n_rows):
        cols = []
        for c in range(n_cols):
            cols.append(round(random.uniform(1, 100), 2))
        rows.append(cols)
    return rows
```

```python
matrix = mat_aleat(10, 7)
matrix
```

### 1.2: A partir de la matriz anterior crea un dataframe de PANDAS, poniendo como nombre de columnas los días de la semana
Si no pudiste crear la matriz del apartado anterior, puedes crear una a partir de NUMPY

```python
import pandas as pd 
import numpy as np

dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]

df = pd.DataFrame(matrix, columns = dias)
df
```

### 1.4: normaliza todos los elementos de la primera matriz. 
Esto es, por cada elemento resta la media y divide por la desviacion tipica del conjunto.

```python
media = np.mean(matrix) 
std = np.std(matrix)
norm = (matrix-media)/std
norm
```

### 1.5: Filtra la matriz normalizada y devuelve los valores entre -1 y 1, sin usar un bucle
El resultado debe ser un vector con dichos valores ordenados descendentemente

```python
mat_filt = (norm[(norm > -1) & (norm < 1)])
mat_filt_sort = np.sort(mat_filt)[::-1]
mat_filt_sort
```

# BLOQUE 2: CREAR HISTOGRAMA Y DIAGRAMA DE BARRAS

### 2.1 Crea un vector de números enteros aleatorios entre el 1 y el 10, de tamaño 100. Crea otro vector de números aleatorios entre el 11 y el 20 del mismo tamaño. 
Muestra por pantalla la tabla de frecuencia de la unión de los 2 vectores para los valores **5,6 y 8 (de la primera)** y para los valores **14,16,19 (de la segunda).**

```python
vect1 = [random.randint(1, 10) for num in range(100)]
vect2 = [random.randint(11, 20) for num in range(100)]
vect = vect1 +vect2

vals = [5, 6, 8, 14, 16, 19]
vect_filt = pd.Series(filter(lambda x: x in vals, vect))
dict_freq = dict(vect_filt.value_counts())
dict_freq
```

### 2.2 crea el histograma para dicho vector de frecuencias 
Con ancho de barra **0.75**, título: **histograma** y color de barra **azul**. Poner un título al **eje x** para diferenciar que son los números que se van a repetir.

```python
import matplotlib.pyplot as plt
def hist(tabla_freq):
    valores = list(tabla_freq.keys())
    frecuencias = list(tabla_freq.values())

    plt.bar(valores, frecuencias, width=0.75, color='blue')

    plt.title("Histogram")
    plt.xlabel("Valores")

    plt.show()
hist(dict_freq)
```

### 2.3 crea un diagrama de barras para el vector fusionado con los 3 elementos más repetidos, sin contar la moda.

```python
moda = max(set(vect), key=vect.count)
vect_moda = pd.Series(vect)[pd.Series(vect) != moda].value_counts().head(3)
hist(dict(vect_moda))
```

# BLOQUE 3: CARGA DE FICHEROS Y MODELIZACION.

Tenemos un fichero llamado **data_examen_sanit** del cual queremos realizar una modelizacion mediante un **arbol de decisión** y una **regresión logística**.
La columna objetivo de la tabla se llama <font color="#9E1B32">**target**</font>, y el resto de columnas actuan como input para predecir el Target

### 3.1 Cargue dicho fichero y prepara los datos para la modelización. Con una proporcion 70-30 para los conjuntos train y test. Obtén los estadisticos más importantes para las variables del conjunto de datos

```python
df = pd.read_csv('data_examen_sanit.txt', delimiter="\t", header=0)
df
```

```python
df.info() # Todas las columnas son numéricas
```

```python
df.describe() # Extraer estadísticas

# Hay varias columnas que más del 75% de los datos son 0, pero no se puede observar si hay variables binarias
```

```python
# Para comprobar cuantas variables son binarias

for col in df.columns:
    uni = df[col].unique()
    if len(uni) == 2:
        print(f"'{col}' es binaria.")
```

```python
df.isna().any().any() # No hay NANs
```

```python
freq = df['target'].value_counts() 
(freq / len(df)) * 100

# Las calses de la variable objetvo están distribuidas en un 66% para la clase 0 y un 33% para la clase 1, approx, por tanto, no se considera un problema imbalanceado
```

```python
# Para observar la correlación entre variables
matriz_corr= df.corr()

mascara = abs(matriz_corr) >= 0.5 # Impongo un filtro para encontrar las variables correlacionadas con mayor facilidad

matriz_corr_filt = matriz_corr[mascara]
matriz_corr_filt
```

```python
# Las variables 10 y 12, 13 y 14 tienen más de 0.75 de correlación, esto es un nivel muy alto y de cara a la modelización se quitará una de ellas (la que tenga menos correlación con el target) para evitar la colinealidad de las variables
# Las variables 11 y 12, 17 y 18, 19 y 20, 21 y 22 tienen una correlación de más de 0.6, pese a ser alta, no se considera suficiente como para tener una colinealidad fuerte y se dejarán todas

# Ninguna tiene una correlación fuerte con la target, solo la var1 supera una correlación de 0.2
```

```python
df1 = df.drop(['var10','var14'], axis = 1)
```

```python
#Partición del dataset

from sklearn.model_selection import train_test_split

x = df1.drop('target', axis=1)  # Variables independientes
y = df1['target'] # Target


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)


freq = y_train.value_counts() 
(freq / len(y_train)) * 100

# Mantiene una proporción simialr al df original
```

```python
from sklearn.preprocessing import StandardScaler

# Como las variables tienen magnitudes muy diferentes, estandarizo los datos
scaler = StandardScaler()

x_train_normalized = scaler.fit_transform(x_train)
x_test_normalized = scaler.transform(x_test)
```

### 3.2 Entrene y evalue los modelos. Arbol y reg. logística

```python
# arbol decision
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier

modelo1 = DecisionTreeClassifier(max_depth=3, min_samples_split=3)  # Por falta de tiempo, no se hace un hyperparameter tuning

modelo1.fit(x_train_normalized, y_train)

y_pred1 = modelo1.predict(x_test_normalized)

y_prob1 = modelo1.predict_proba(x_test)[:, 1]


confusion_matrix = confusion_matrix(y_test, y_pred1)

# Tiene una tasa de falsos negativos aun muy alta

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

```python
#regresión logística

from sklearn.linear_model import LogisticRegression

modelo2 = LogisticRegression()

modelo2.fit(x_train_normalized, y_train)

y_pred2 = modelo2.predict(x_test_normalized)


y_prob2 = modelo2.predict_proba(x_test)[:, 1] 

confusion_matrix = confusion_matrix(y_test, y_pred2)

# Tiene una tasa de falsos negativos aun más alta que el caso del árbol

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

### 3.3 Proponga un método para comparar ambos modelos (ej: Curva ROC)

```python
from sklearn.metrics import accuracy_score

# Al no ser un problema imbalanceado, se podría utilizar la accuracy

acc1 = accuracy_score(y_test, y_pred1)
acc2 = accuracy_score(y_test, y_pred2)
print(f"Accuracy arbol: {acc1}")
print(f"Accuracy logistica: {acc2}")
```

```python
#Utiliando la curva ROC:
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_test, y_prob1)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC arbol')
plt.legend(loc='lower right')
plt.show()
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob2)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC logistica')
plt.legend(loc='lower right')
plt.show()
```

```python
# Los resutlados son malos en ambos modelos, al haber observado la matriz de confusión y la curva ROC, se observa que la logistica encuentra mejores resultados,

# EL accuracy nos habría llevado a una conclusión distinta, ya que solo tiene en cuenta los aciertos TP y FN
```
